```python
import pandas as pd
tabela = pd.read_csv("telecom_users.csv")

tabela = tabela.drop("Unnamed: 0", axis=1) #duas informações nome da coluna na primeira ou linha e na segunda se for linha axis=0 se for coluna axis=1 


tabela["TotalGasto"] = pd.to_numeric(tabela["TotalGasto"], errors="coerce") # passa duas informações a coluna e a segunda os erros que serão ignorados
#para selecionar as colunas de uma tabela sempre usa colchetes
#sempre exclua as colunas antes das linhas 


tabela = tabela.dropna(how="any",axis=1) 
#exclui valores vazios how= e axis= | all quando exclui colunas completamente vazias 
# any é quando voce quer excluir colunas que tenha pelo menos 1 valor vazio

#contar os valores da coluna churn

#display(tabela["ServicoInternet"].value_counts())
display(tabela["ServicoInternet"].value_counts(normalize=True).map())
#display(tabela)

#analise de dados com plotly(Graficos)
```
