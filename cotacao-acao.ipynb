```python
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt



#importando a base de dados em excel
tabela_empresas = pd.read_excel("Empresas.xlsx")
#printando a base de dados do excel
display(tabela_empresas)

#leitura da base de dados da coluna empresa e a informação do valor de fechamento Adj Close na respectiva data de cada empresa da tabela e mostrando em grafico tambem
for empresas in tabela_empresas['Empresas'] :
    print(empresas)
    cotacao_bovespa = web.DataReader(f'{empresas}.SA', data_source='yahoo', start="01/01/2021", end="08/14/2021")
    cotacao_bovespa["Adj Close"].plot(figsize=(15,10))
    display(cotacao_bovespa)
    plt.show()

```
